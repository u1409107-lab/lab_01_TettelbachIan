## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

## Preview Data

Use the following data dictionary to understand the attributes available


# Create basic charts

```python
#Line chart
df.plot(x='carmodel', y='mpg', kind='line', title="MPG by Car")
```

```python
#Boxplot
df['mpg'].plot(kind='box', title = 'Miles Per Gallon')
```

```python
#Scatter plot
df.plot.scatter(x='mpg',y='hp',c='wt')
```

- 1.1 Create a pie chart showing the count of cars by transmission type.  Display labels showing the percentage of each transmission type.  
- 1.2 Create a bar chart of cars and HP, with different colors for each car model
- 1.3 Create a scatter plot of wt (weight) vs. mpg.
- 1.4 Create a bar chart showing the average mpg for each cyl.

```python
#Pie chart ðŸ”§
transmission_counts = df['am'].value_counts()
transmission_labels = {0: 'Automatic', 1: 'Manual'}
transmission_counts.index = transmission_counts.index.map(transmission_labels)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
transmission_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Car Count by Transmission Type')
plt.ylabel('') # Hide the default 'am' label
plt.show()
```

```python
#Bar chart ðŸ”§
import matplotlib.pyplot as plt

# Create the bar chart
# Generate a list of colors, one for each car model
num_cars = len(df)
colors = plt.cm.get_cmap('tab20', num_cars) # Use a colormap for distinct colors

plt.figure(figsize=(15, 6)) # Adjust figure size for better readability
plt.bar(df['carmodel'], df['hp'], color=[colors(i) for i in range(num_cars)])
plt.xlabel('Car Model')
plt.ylabel('Horsepower (HP)')
plt.title('Horsepower by Car Model')
plt.xticks(rotation=90, fontsize=8) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
#Scatter plot ðŸ”§
df.plot.scatter(x='wt', y='mpg', title='Weight vs. MPG')
```

```python
#Bar chart ðŸ”§
df.groupby('cyl')['mpg'].mean().plot(kind='bar', title='Average MPG by Number of Cylinders', ylabel='Average MPG')
```

```python
#Histogram ðŸ”§
df['wt'].plot(kind='hist', bins=10, title="Weight Distribution")
```
